```python
import scipy.stats as stats
import random
import statistics
import pandas as pd
import os
```

```python
import os 
import pandas as pd 

input_folder = './'
output_file = './merged.txt' 

data_frames = []

for filename in os.listdir(input_folder): 
    if filename.endswith('.tsv'): 
        base_name = os.path.splitext(filename)[0]
        prefix = base_name.split('.IS',  1)[0]
        
        file_path = os.path.join(input_folder,  filename)
        df = pd.read_csv(file_path,  sep='\t')
        df.insert(0,  'sample', prefix)
        
        data_frames.append(df) 
 
merged_data = pd.concat(data_frames,  ignore_index=True)
merged_data.to_csv(output_file,  sep='\t', index=False)
 
 
```

```python
output_file = './merged_breadthFiltered.txt'

filtered_data = merged_data[merged_data['breadth'].astype(float) >= 0.25]

filtered_data.to_csv(output_file,  sep='\t', index=False)
```

```python
MAG_dictionary = {}

for line in open("merged_breadthFiltered.txt", "r"):
    if line.startswith("sample"):
        continue
    else:
        line = line.strip().split("\t")
        sample = line[0]
        MAG = line[1]
        if MAG not in MAG_dictionary.keys():
            MAG_dictionary[MAG] = sample
        else:
            MAG_dictionary[MAG] += "\t" + sample
```

```python
MAG_dictionary_path = "MAGpresence.txt"
MAG_dictionary_file = open(MAG_dictionary_path, "w")

for key, value in MAG_dictionary.items():
    print(key + "\t" + value, file = MAG_dictionary_file)
    
MAG_dictionary_file.close()
```

```python
res_wtp_counts = {}
total_wtp_res = []
for fa_file, contents in MAG_dictionary_diff.items():
    entries = contents.split('\t')
    res_types = set()
    for entry in entries:
        parts = entry.split('_')
        if len(parts) > 2 and parts[-2] == 'RES':
            res_type = parts[-1]
            res_types.add(res_type)
    
    for res in res_types:
        wtp_count = []
        for entry in entries:
            parts = entry.split('_')    
            if len(parts) > 2 and parts[-2] == 'WTP':
                wtp_type = parts[-1]
                if wtp_type == res:
                    wtp_count.append(wtp_type)
        wtp_count_set = set(wtp_count)
        res_count = []
        for entry in entries:
            parts = entry.split('_')   
            if len(parts) > 2 and parts[-2] == 'RES':
                res_type_diff = parts[-1]
                if res_type_diff == res:
                    res_count.append(res_type_diff)
        total_wtp_res.append(len(wtp_count) * len(res_count))

print(f"WTP-RES pair of same site: {sum(total_wtp_res)}")
```
